```python
# Group by 'Item' and sum the 'Calories' for each item
calories_by_item = df.groupby('Item')['Calories'].sum().reset_index()
calories_by_item_sorted = calories_by_item.sort_values(by='Calories', ascending=False)
calories_by_item_sorted
```

```python
# Visualize the distribution of Total Fat, Carbohydrates, and Protein
fig, ax = plt.subplots(3, 1, figsize=(12, 18))

sns.histplot(df['Total Fat'], kde=True, bins=30, color='green', ax=ax[0])
ax[0].set_title('Total Fat Distribution')
ax[0].set_xlabel('Total Fat (g)')
ax[0].set_ylabel('Frequency')

sns.histplot(df['Carbohydrates'], kde=True, bins=30, color='orange', ax=ax[1])
ax[1].set_title('Carbohydrates Distribution')
ax[1].set_xlabel('Carbohydrates (g)')
ax[1].set_ylabel('Frequency')

sns.histplot(df['Protein'], kde=True, bins=30, color='red', ax=ax[2])
ax[2].set_title('Protein Distribution')
ax[2].set_xlabel('Protein (g)')
ax[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Histogram of Calories
plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], kde=True, bins=30, color='blue')
plt.title('Calories Distribution Across McDonald\'s Menu Items')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for Calories by Category
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Calories', data=df)
plt.title('Calorie Distribution by Category')
plt.xlabel('Food Category')
plt.ylabel('Calories')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Boxplot for Total Fat by Category
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Total Fat', data=df)
plt.title('Total Fat Distribution by Category')
plt.xlabel('Food Category')
plt.ylabel('Total Fat (g)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Boxplot for Protein by Category
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Protein', data=df)
plt.title('Protein Distribution by Category')
plt.xlabel('Food Category')
plt.ylabel('Protein (g)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
avg_nutrition = df.groupby('Category')[['Calories', 'Total Fat', 'Protein', 'Carbohydrates']].mean().reset_index()
avg_nutrition
```

```python
# Bar chart of average Calories by Category
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Calories', data=avg_nutrition, hue='Category', palette='viridis', legend=False)
plt.title('Average Calories by Food Category')
plt.xlabel('Food Category')
plt.ylabel('Average Calories')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Introduction

```python
#The objective of this analysis was to explore the nutritional content of McDonald's menu items, focusing on key metrics like calories, fat, protein, 
#and carbohydrates. By analyzing the dataset, I aimed to provide insights that could guide both McDonald’s customers and the organization itself. These 
#insights help in making informed choices about menu options and can also guide improvements in nutritional offerings.
```

Findings and Insights from the Analysis

```python
#A. Highest and Lowest Calorie Menu Items
#Top 10 Highest Calorie Items: These typically included high-fat and high-protein items such as the Chicken & Fish, Quarter Pounder, and Double Cheeseburger. 
#These menu items often contain a high amount of total fat and carbohydrates, contributing to their high calorie content.

#Top 10 Lowest Calorie Items: These items included lighter options like Side Salad, Apple Slices, and lighter drinks such as Iced Coffee. 
#These options have significantly fewer calories, fat, and carbohydrates, making them better choices for customers who are calorie-conscious or following
#specific dietary restrictions.

#B. Average Nutritional Content by Category
#By grouping items into categories such as Burgers, Salads, and Desserts, the following insights were obtained:

#Burgers: Generally, burgers had high calories (average ~500), high fat (~25g), and moderate protein (~27g). These items are calorie-dense and typically 
#high in fat, making them less ideal for health-conscious customers unless consumed in moderation.

#Salads: Salads showed a much lower calorie count (average ~150), lower fat (~5g), and higher fiber content. They are a healthier alternative compared to
#burgers, offering a lighter option with fewer calories and fat while providing essential nutrients like vitamins.

#Desserts: Desserts, on average, were higher in carbohydrates (sugars), but they often contained lower amounts of protein and fat. The average calorie 
#content for desserts was around 300. While tasty, desserts contribute significantly to sugar intake and are less nutrient-dense compared to other 
#categories.
```

Recommendations for McDonald's Customers

```python
# Healthy Options:
#Salads and Fruit Options: For customers seeking lower-calorie and nutritious meals, salads and fruit-based options are excellent choices. These items 
#are generally lower in calories, fat, and sodium.

#Smaller Portion Sizes: Offering smaller portion sizes or reduced-calorie options for high-calorie items like burgers and sandwiches could cater to 
#health-conscious customers.
    
#Protein-rich, Low-Carb Meals: For customers on low-carb diets, items like grilled chicken salads or wraps offer a good balance of protein with fewer 
#carbs.
    
#Less Healthy Options:
#High-Calorie Burgers and Fries: McDonald's could consider reformulating some of their higher-calorie items, such as large burgers with fries, by 
#reducing portion sizes, using lower-fat cooking methods, or offering healthier alternatives (e.g., baked fries).
    
#Sodium Reduction: Reducing the sodium content in high-calorie menu items, particularly in burgers and sandwiches, could make the menu healthier overall.
#Items like breakfast sandwiches and fried foods tend to have higher sodium content.
    
#Customized Nutritional Information:
#McDonald's can offer a calorie and nutritional information filter on their menu, allowing customers to select meals that align with their dietary needs,
#whether it’s lower calorie, low sodium, or high protein.
```

Nutrition-Based Recommendations for McDonald's

```python
#Based on the analysis, we can make several recommendations for McDonald's:

#Offer More Low-Calorie Options: McDonald's could introduce more low-calorie menu items with healthier ingredients, such as more plant-based options or 
#grilled protein choices with fewer calories and less fat.

#Reduce Portion Sizes: Many high-calorie items come in large portions. Offering smaller portion sizes or smaller meal options could help customers control
#their calorie intake.

#Improve Menu Labeling: McDonald’s could benefit from providing clearer nutritional information on their menu and in-store displays, such as highlighting
#healthier options or offering calorie counts for different sizes of the same meal.

#Increase Healthy Sides: Expanding healthier side options (e.g., salads, fruits, and low-fat dressing options) could encourage customers to make better 
#choices while keeping the overall meal balanced.

#Promote Low-Fat and Low-Sugar Items: By promoting items like salads or grilled chicken sandwiches as part of a "health-conscious" menu, McDonald's can 
#cater to a growing demand for healthier fast food.

#Limit High Sodium Options: Some items exceed the recommended daily sodium intake. Reducing sodium in items like burgers and fries would help improve the
#overall healthiness of the menu.
```
